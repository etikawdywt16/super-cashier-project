```python
from feature import *
```

```python
def menu():
    """This function is menu of Super Cashier system."""
    
    user = Transaction()
    
    # Run this code if all input is valid
    try:
        # This is main menu
        while True:
            print("")
            print(" "*15, "Menu", " "*15)
            print("="*36)
            print("1. Add item to the shopping list")
            print("2. Update item on the shopping list")
            print("3. Remove item on the shopping list")
            print("4. Check order on the shopping list")
            print("5. Reset the shopping list")
            print("6. Total price")
            print("7. Cancel transaction")

            print("")
            choice = int(input("Choice: "))

            if choice == 1:
                user.add_item()
            elif choice == 2:

                # This is menu to update name item, quantity item, and price item
                while True:
                    print("")
                    print(" "*15, "Menu", " "*15)
                    print("="*36)
                    print("1. Update item name")
                    print("2. Update item quantity")
                    print("3. Update item price")
                    print("4. Back to main menu")

                    print("")
                    choice_2 = int(input("Choice: "))

                    if choice_2 == 1:
                        user.update_item_name()
                    elif choice_2 == 2:
                        user.update_item_qty()
                    elif choice_2 == 3:
                        user.update_item_price()
                    elif choice_2 == 4:
                        break
                    else:
                        print("Incorrect choice")

            elif choice == 3:
                user.delete_item()
            elif choice == 4:
                user.check_order()
            elif choice == 5:
                
                # This is check-out confirmation
                while True:
                    print("")
                    print("Are you sure want to reset the shopping list? All items will be removed from the shopping list")
                    print("1. Yes")
                    print("2. No")

                    print("")
                    choice_3 = int(input("Choice: "))
                    print("")

                    if choice_3 == 1:
                        user.reset_item()
                        break
                    elif choice_3 == 2:
                        break
                    else:
                        print("Incorrect choice")
                                      
            elif choice == 6:
                user.total_price()

                # This is menu to check out all items on the shopping list
                while True:
                    print("")
                    print(" "*15, "Menu", " "*15)
                    print("="*36)
                    print("1. Check out")
                    print("2. Back to main menu")

                    print("")
                    choice_4 = int(input("Choice: "))

                    if choice_4 == 1:

                        # This is check-out confirmation
                        while True:
                            print("")
                            print("Are you sure want to check out all items on shopping list?")
                            print("1. Yes")
                            print("2. No")

                            print("")
                            choice_5 = int(input("Choice: "))
                            print("")

                            if choice_5 == 1:
                                user.check_out()
                                return
                            elif choice_5 == 2:
                                break
                            else:
                                print("Incorrect choice")

                    elif choice_4 == 2:
                        break
                    else:
                        print("Incorrect choice")

            elif choice == 7:

                # This is cancel transaction confirmation
                while True:
                    print("")
                    print("Are you sure want to cancel transaction? All items will be removed from the shopping list.")
                    print("1. Yes")
                    print("2. No")

                    print("")
                    choice_6 = int(input("Choice: "))

                    if choice_6 == 1:
                        user.reset_item()
                        user.cancel()
                        return
                    elif choice_6 == 2:
                        break
                    else:
                        print("Incorrect choice")

            else:
                print("Incorrect choice")
    
    # Run this code if the input is invalid
    except ValueError:
        print("")
        print("Incorrect input")
        menu()
```
